```python
from urllib.request import Request, urlopen, urlretrieve
from bs4 import BeautifulSoup
import time
import requests
import random
```

```python
"""Returns a list of image URLs from a given shot type url

Args:
    baseurl (str): The URL of the shot type 
    print_cols (int): Number of pages to scrape from

Returns:
    url_links: a list of URLs to all images of the desired shot type
"""
def scrapeFilmGrab(baseurl):
    url_links = []
    #url = baseurl + str(i)
    #print(url)
    req = Request(baseurl, headers={'User-Agent': "Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 \
                                    (KHTML, like Gecko) Chrome/41.0.2272.0 Safari/537.36"})
    page = urlopen(req).read()
    #req = requests.get(baseurl, headers={'User-Agent': 'Mozilla/5.0'})
    page_soup = BeautifulSoup(page, "html.parser")
    articles = page_soup.findAll("li", {"class": "listing-item"})
    for article in articles:
        urllink = article.find('a').get('href')
        url_links.append(urllink)
    return url_links
```

```python
def scrapeFilmGrabYears(baseurl, numPages):
    url_links = []
    for i in range(1, numPages+1):
        url = baseurl + str(i)
        #print(url)
        req = Request(url, headers={'User-Agent': "Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 \
                                    (KHTML, like Gecko) Chrome/41.0.2272.0 Safari/537.36"})
        page = urlopen(req).read()
        page_soup = BeautifulSoup(page, "html.parser")
        articles = page_soup.find_all("a", {"class": "popup-image"})
        for article in articles:
            urllink = article.get('href')
            url_links.append(urllink)
        time.sleep(0.5)
        
    return url_links
```

```python
def scrapeImages(baseurl):
    img_links = []
    #url = baseurl + str(i)
    #print(url)
    req = Request(baseurl, headers={'User-Agent': "Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 \
                                    (KHTML, like Gecko) Chrome/41.0.2272.0 Safari/537.36"})
    page = urlopen(req).read()
    #req = requests.get(baseurl, headers={'User-Agent': 'Mozilla/5.0'})
    page_soup = BeautifulSoup(page, "html.parser")
    images = page_soup.findAll("div", {"class": "bwg-item"})
    #counter = 0
    for image in images:
        #if (counter % 2 == 0):
        imglink = image.find('a').get('href')
        img_links.append(imglink)
        #counter+=1

    return img_links
```

```python
import requests # to get image from the web
import shutil # to save it locally

## Set up the image URL and filename
def saveImages(urlArray, filepath):
    for image_url in urlArray:
        filename = image_url.split("/")[-1]

        # Open the url image, set stream to True, this will return the stream content.
        r = requests.get(image_url, headers={'User-Agent': "Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 \
                                    (KHTML, like Gecko) Chrome/41.0.2272.0 Safari/537.36"}, stream = True)
        # Check if the image was retrieved successfully
        if r.status_code == 200:
            # Set decode_content value to True, otherwise the downloaded image file's size will be zero.
            r.raw.decode_content = True

            # Open a local file with wb ( write binary ) permission.
            with open(filepath+filename,'wb') as f:
                shutil.copyfileobj(r.raw, f)
        else:
            print(image_url)
        time.sleep(0.5)
```
